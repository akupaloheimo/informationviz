```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from shapely.geometry import Point
```

```python
# Load data
trips = pd.read_csv("/home/akupalo/informationviz/data/2021-04.csv")
stations = pd.read_csv("/home/akupalo/informationviz/data/Helsingin_ja_Espoon_kaupunkipyöräasemat_avoin.csv")

# Ensure matching column names
stations.rename(columns={"ID": "Departure station id"}, inplace=True)

# Count departures per station
departure_counts = trips["Departure station id"].value_counts().reset_index()
departure_counts.columns = ["Departure station id", "Departure Count"]

# Merge with station coordinates
merged_df = pd.merge(stations, departure_counts, on="Departure station id", how="left")
merged_df.fillna(0, inplace=True) 

# Convert to GeoDataFrame
geometry = [Point(xy) for xy in zip(merged_df["x"], merged_df["y"])]
stations_gdf = gpd.GeoDataFrame(merged_df, geometry=geometry, crs="EPSG:4326")

# Get station coordinate boundaries
min_lon, max_lon = stations_gdf.geometry.x.min(), stations_gdf.geometry.x.max()
min_lat, max_lat = stations_gdf.geometry.y.min(), stations_gdf.geometry.y.max()

print(min_lon)
print(max_lon)
print(min_lat)
print(max_lat)

# Load Helsinki map image
map_path = "/home/akupalo/informationviz/data/Map_of_Helsinki.png"
map_img = mpimg.imread(map_path)
map_extent = [24.70, 25.16, 60.08, 60.33]

# Create the plot
fig, ax = plt.subplots(figsize=(10, 11))
ax.imshow(map_img, extent=map_extent, aspect='auto')  # Display the map as background

# Plot bike stations with sizes based on departure counts
stations_gdf.plot(
    ax=ax,
    markersize=stations_gdf["Departure Count"] / 15,  # Scale dot size
    alpha=0.6, color="blue", edgecolor="black"
)

# Labels and title
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_title("Bike Departures by Station in Helsinki (April 2021)")

# Show the plot
plt.show()
```
