```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx

# Read the data
node_df = pd.read_csv("/home/akupalo/informationviz/data/sdg_network.csv")
edge_df = pd.read_csv("/home/akupalo/informationviz/data/sdg_connections.csv")

# Create the graph
G = nx.Graph()

# Add nodes with color attributes
for _, row in node_df.iterrows():
    G.add_node(row['SDG_ID'], color=row['color'], name=row['SDG_name'])

# Add edges with weight attributes
for _, row in edge_df.iterrows():
    G.add_edge(row['sdg1'], row['sdg2'], weight=row['weight'])

# Create a color map from the node colors
node_colors = [data['color'] for _, data in G.nodes(data=True)]

# Draw the graph with a smaller figure size
fig, ax = plt.subplots(figsize=(8, 8))  # Reduced from 12x12 to 8x8

# Use circular layout with reduced scale
pos = nx.circular_layout(G, scale=0.8) 

# Draw nodes with smaller size
nx.draw_networkx_nodes(
    G, 
    pos, 
    node_color=node_colors,
    node_size=500,  # Reduced from 1000 to 500
    ax=ax
)

# Draw labels with smaller font
nx.draw_networkx_labels(
    G, 
    pos, 
    labels={n: n for n in G.nodes()},
    font_size=10,  # Reduced from 12 to 10
    font_color='white',
    font_weight='bold',
    ax=ax
)

# Draw edges with thinner lines
nx.draw_networkx_edges(
    G, 
    pos, 
    width=[d['weight'] for _, _, d in G.edges(data=True)],  # Thinner edges
    edge_color='gray',
    alpha=0.7,
    ax=ax
)
plt.title('SDG connections')
ax.set_axis_off()
plt.tight_layout()
plt.show()
```

```python
import random

# Initialize random positions (simulates a seed)
random.seed(42)  # Fix randomness for reproducibility
initial_pos = {node: (random.random(), random.random()) for node in G.nodes()}

# Draw the graph with a smaller figure size
fig, ax = plt.subplots(figsize=(8, 8))

# Use circular layout with reduced scale
pos = nx.kamada_kawai_layout(G, scale=0.8, pos=initial_pos) 

# Draw nodes with smaller size
nx.draw_networkx_nodes(
    G, 
    pos, 
    node_color=node_colors,
    node_size=500,  # Reduced from 1000 to 500
    ax=ax
)

# Draw labels with smaller font
nx.draw_networkx_labels(
    G, 
    pos, 
    labels={n: n for n in G.nodes()},
    font_size=10,  # Reduced from 12 to 10
    font_color='white',
    font_weight='bold',
    ax=ax
)

# Draw edges with thinner lines
nx.draw_networkx_edges(
    G, 
    pos, 
    width=[d['weight'] for _, _, d in G.edges(data=True)],  # Thinner edges
    edge_color='gray',
    alpha=0.7,
    ax=ax
)

ax.set_axis_off()
plt.tight_layout()
plt.show()
```
